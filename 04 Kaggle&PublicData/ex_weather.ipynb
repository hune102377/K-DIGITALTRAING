# 대구가 가장 더웠던 날은?
 - 1. csv 데이터를 읽어옴
 - 2. 순차적으로 최고 기온을 확인 (비교)
 - 3. 최고 기온이 가장 높았던 날짜의 데이터를 저장
     - 문자열 형태의 데이터를 실수 형태로 변환
     - 지금까지의 최고 기온 값보다 현재 행(row)의 최고 기온이 더 높으면
         - 최고 기온 날짜 업데이트
         - 최고 기온 값 업데이트
 - 4. 최종 저장된 데이터 출력

```python
# 대구 최고 기온 날짜와 최고 온도 구하기
import csv

f = open(DIR_PATH + FILENAME, 'r', encoding='euc_kr')
data = csv.reader(f, delimiter=',')

header = next(data)
max_temp = -999 # 최고 기온을 저장할 변수 초기화
max_date = '' # 최고 기온의 날짜를 저장할 변수 초기화

for row in data:
    if row[-1] == '': # [-1]: 리스트에서 마지막 데이터가 없는 경우
        row[-1] = -999
    row[-1] = float(row[-1])
    
    if(max_temp < row[-1]):
        max_temp = row[-1]
        max_date = row[0] #  날짜 : index[0]
        
f.close()

print(f'기상 관측 이래 대구의 최고 기온이 가장 높았던 날은 {max_date}로 {max_temp}도 였습니다.')
```

```python
# Numpy를 활용한 대구 최고 기온 날짜와 최고 온도 구하기
import csv
import numpy as np

f = open(DIR_PATH + FILENAME, 'r', encoding='euc_kr')
data = csv.reader(f, delimiter=',')

header = next(data)
max_temp = -999 # 최고 기온을 저장할 변수 초기화
max_date = '' # 최고 기온의 날짜를 저장할 변수 초기화
max_temp_list = list() # 최고 기온을 저장할 리스트
max_temp_date = list() # 날짜 정보를 저장할 리스트

for row in data:
    if row[-1] == '': # [-1]: 리스트에서 마지막 데이터가 없는 경우
        row[-1] = -999
    
    max_temp_list.append(row[-1])
    max_temp_date.append(row[0])
        
f.close()

# 리스트를 ndarray로 변환 -> 문자열을 실수로 변환(astype)
max_temp_array = np.array(max_temp_list) # 리스트를 ndarray 타입으로 변경
max_temp_array = max_temp_array.astype(float) # 문자열 타입을 float로 변경
max_date_array = np.array(max_temp_date) # 리스트를 nadarray 타입으로 변경

#  실제 값을 구하는 부분 : max()로 최대값 추출, 최대값의 인덱스를 argmax()로 추출
max_temp = max_temp_array.max()
index = max_temp_array.argmax() #  최대값의 index 리턴
max_date = max_date_array[index]

print(f'기상 관측 이래 대구의 최고 기온이 가장 높았던 날은 {max_date}로 {max_temp}도 였습니다.')
```

```python
# 8월만 추출하기
# Numpy를 활용한 대구 최고 기온 날짜와 최고 온도 구하기
import csv
import matplotlib.pyplot as plt

f = open(DIR_PATH + FILENAME, 'r', encoding='euc_kr')
data = csv.reader(f, delimiter=',')
next(data)
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if '8' in month :
            aug.append(float(row[-1]))
        
f.close()

plt.hist(aug, bins = 100, color = 'tomato')
plt.xlabel("Temperature")
plt.ylabel("Counts")
plt.show()
```

```python
# 1월과 8월의 기온 데이터 히스토그램
import csv
import matplotlib.pyplot as plt

f = open(DIR_PATH + FILENAME, encoding='euc_kr')
data = csv.reader(f)
next(data) # 데이터에서 첫 줄은 넘김
aug = []
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))

f.close()
plt.hist(aug, bins = 100, color = 'tomato', label='Aug') # 히스토그램(도수분포표)로 표시 bins = 가로축 수
plt.hist(jan, bins = 100, color = 'b', label='Jan')
plt.xlabel("Temperature")
plt.legend() # 범례를 추가
plt.rc('axes', unicode_minus = False) # - 기호 깨지는 것을 방지
plt.show()
```

```python
# 특정 날짜의 최고 기온 찾기
import csv
import matplotlib.pyplot as plt

def draw_graph_on_date(month, day):
    f = open(DIR_PATH + FILENAME, encoding='euc_kr')
    data = csv.reader(f)
    next(data)
    result = []
    for row in data:
        if row[-1] != '':
            if row[-1] != '':
                date_string = row[0].split('-')
            if date_string[1] == month and date_string[2] == day:
                result.append(float(row[-1]))
    f.close()
    plt.figure(figsize=(15,2))
    plt.plot(result, 'royalblue')
    plt.show()
    
month, date = input('날짜(월 일)를 입력하세요: ').split()

draw_graph_on_date(month, date)
```
